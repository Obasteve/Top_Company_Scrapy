```python
from urllib.request import urlopen as uReq
```

```python
#The website to scrape is passed into the urls
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
```

```python
#Reading out the html page of the website to scrape
```

```python
uclient = uReq(url)
page_html=uclient.read()
page_soup= soup(page_html, 'html.parser')

```

```python
# The table of interest is called out
```

```python
page_soup.find('table')
```

```python
page_soup.find_all('table')[0]
```

```python
#Find out the class of the table tag
```

```python
page_soup.find('table', class_='wikitable sortable sticky-header-multi sort-under')
```

```python
page_soup.find('th', class_='headerSort')
```

```python
table=page_soup.find_all('table')[0]
table
```

```python
# We bring out the headers of the table to scrape
```

```python
# Find all header titles
world_titles = page_soup.find_all('th')

# Extract and clean the text from each header title
world_table_titles = [title.text.strip() for title in world_titles[0:9]]

# Print the first 7 titles as a horizontal list
print(", ".join(world_table_titles))
```

```python
#Header is saved in a pandas datatframe
```

```python
import pandas as pd

df= pd.DataFrame(columns=world_table_titles)
df
```

```python
table=page_soup.find_all('table')[0]
```

```python
#We write a loop that calls all "tr" tag the from the html table tag and the pass it to the dataframe header
```

```python
column_data=table.find_all('tr')
```

```python

for row in column_data[2:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    # Check if the number of columns matches the DataFrame
    if len(individual_row_data) == len(df.columns):
        length = len(df)
        df.loc[length] = individual_row_data
    else:
        print(f"Skipping row with mismatched columns: {individual_row_data}")
        print(f"Expected columns: {len(df.columns)}, but got {len(individual_row_data)}")
```

```python
df
```
